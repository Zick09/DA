```python
import requests
import json
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
# извлекаем вакансии по работе с данными в г.Краснодар
page = 0
num_per_page = 100
city = 53
text = 'data'
url = f'https://api.hh.ru/vacancies?text={text}&page={page}&per_page={num_per_page}&area={city}&only_with_salary=true'
```

```python
res = requests.get(url)
vacancies = res.json()
num_pages = vacancies.get('pages')
vacancies_ids = [el.get('id') for el in vacancies.get('items')]
```

```python
vacancies.get('items')
```

```python
# создаем пустой датафрейм
col = ['id','name','salary','published_at','responsibility']
df = pd.DataFrame(columns = col)
```

```python
# построение таблицы с описанием
for i in vacancies['items']:
    temp = list(i.values())
    ser = [temp[0],temp[2],temp[7],temp[13], temp[21]['responsibility']]
    temp_df = pd.DataFrame([ser], columns = col)
    df = df.append(temp_df,ignore_index=True)
```

```python
# переведем даты в формат datetime
df.published_at = pd.to_datetime(df.published_at)
```
